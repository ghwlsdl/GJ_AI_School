# [제2차 조별 과제]

우리반 학우들 중에 온라인 반 학우 3명, 오프라인 반 학우 3명의 이름을 서로 물어보아 알아내서,

그 데이터를 바탕으로 리스트, 튜플, 딕셔너리 형태로 표현하고,

그 결과를 출력하는 프로그램을 작성하시오.

(단, 출력 모양에는 구애 받지 않아도 됩니다.)

# [계획]
1. 온라인 반 학우 & 오프라인 반 학우로 리스트, 튜플, 딕셔너리를 위한 변수를 지정해 줍니다.
2. 변수가 너무 길어질 경우엔, 변수를 재지정해 줍니다.
3. 출력해봅니다.

# [1차 코딩]

계획에 따라서 진행을 진행해보았습니다.

```python
# 온라인 학생 list, tuple, set, dic
online_student_list = ['김재은', '하지성', '최아름']
online_student_tuple = ('김재은', '하지성', '최아름')
online_student_dic = {'김재은' : '온라인' , '하지성' : '온라인', '최아름': '온라인'}

# 오프라인 학생 list, tuple, set, dic
offline_student_list = ['주성민', '박지호', '이찬영']
offline_student_tuple = ('주성민', '박지호', '이찬영')
offline_student_dic = {'주성민' : '오프라인', '박지호' : '오프라인', '이찬영' : '오프라인'}

# 변수 단순화 진행
onsl = online_student_list
onst = online_student_tuple
onsd = offline_student_dic

offsl = offline_student_list
offst = offline_student_tuple
offsd = online_student_dic

print(onsl, onst, onsd, offsl, offst, offsd)
```

# [1차 수정]

print()를 1열로 나열되는 것은 가독성이 떨어지므로

하나의 출력이 나오고 난 뒤에 줄 바꿈이 이루어지도록 수정해 보았습니다.

```python
print(onsl, onst, onsd, offsl, offst, offsd, sep='\n')
```

# [2차 수정]

### time 라이브러리를 사용해서 출력까지의 소요 시간을 확인해보았습니다.

```python
from time import time, sleep

start = time()

online_student_list = ['김재은', '하지성', '최아름']
online_student_tuple = ('김재은', '하지성', '최아름')
online_student_dic = {'김재은' : '온라인' , '하지성' : '온라인', '최아름': '온라인'}

offline_student_list = ['주성민', '박지호', '이찬영']
offline_student_tuple = ('주성민', '박지호', '이찬영')
offline_student_dic = {'주성민' : '오프라인', '박지호' : '오프라인', '이찬영' : '오프라인'}

onsl = online_student_list
onst = online_student_tuple
onsd = online_student_dic

offsl = offline_student_list
offst = offline_student_tuple
offsd = offline_student_dic

print(onsl, onst, onsd, offsl, offst, offsd, sep='\n')

end = time()

print('time elapsed:', end - start)
```

> [(온) 강은석의 리뷰]
* time을 사용해서 속도개선이 이루어졌는지 확인하는 습관을 가져야 겠다는 생각을 하게 되었습니다. 잘 배웠습니다!

> [(오)오준묵의 리뷰]
 * 프로그래밍에서 변수가 길어질 경우를 대비한 것이 인상적이었다.
 * 위의 강은석 학우님의 리뷰처럼 time을 이용한 속도개선 방법을 배울 수 있어서 좋았다.

> [(오) 이재혁의 리뷰]
* 변수가 길어질 때, 어떻게 하면 간소화하고 가독성이 좋게 만들 수 있는지 배울 수 있는 좋은 코딩스타일이었습니다.
