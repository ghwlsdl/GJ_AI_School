### 1. 클래스 정의

```python
class Seven_NLP():

    def __init__(self):
      self.number_of_people = 0
      self.lecturer = "박진수"
      self.teacher = "최영화"
      self.equipment_for_deep_learning = "Laptop_with_high-end_GPU"
      self.goal = "Let's develop AI development capabilities and get a job at the company you want!"

    def get_main_teacher_name(self):
        return self.lecturer

    def get_teacher_name(self):
        return self.teacher

    def get_equipment_for_deep_learning(self):
        return self.equipment_for_deep_learning

    def student_goal(self):
        return self.goal

    def main_teacher(self):
        return "Did you prepare well for the lecture?"

    def student(self):
        return "To study AI at.광주인공지능사관학교"

    def study_site(self, site):
      if site == "classroom":
        return "Off-line"
      if site == "home":
        return "On-line"

    def attention_degree(self, level):
      self.level = level
        return self.level
    
    def attention_change(self):
      if self.level >= 50:
        self.level -= 10
        return "active Coding"      
      else:
        self.level +=10
        return "need rest"

    def laptop_gpu(self, rtx_number):
      
      self.rtx_number = rtx_number

      if rtx_number > 3060:
        return "Good"
      else:
        return "Not Enough"
```

### 2. 인스턴스 생성

```python
sn = Seven_NLP()
```
