> **[과제 내용]**
* 여러분이 만든 클래스에서 데이터를 받을 수 있게 해 보시오. 
* 여러분이 만든 클래스를 가지고 데이터를 여러 모양으로 넣어서 여러 종류의 인스턴스를 만들어 보시오.


```python
class GjaiClass():

    def __init__(self,
                 lecturer_name = "",
                 teacher_name = "",
                 phone_number = "",
                 student_num = 0,
                 class_subject = ""
                ):
      
        self.__lecturer = lecturer_name
        self.__lecturer_num = phone_number
        self._teacher = teacher_name
        self.student_num = student_num
        self.class_subject = class_subject


    # 은닉된 변수의 값을 불러내는 함수 (게터)
    def get_hided_information(self, title):
        if title == "name":
            print(self.__lecturer)
        elif title == "number":
            print(self.__lecturer_num)
        else:
            print("error")

    # 은닉된 함수
    def __get_test_score(self):
        self.score = [50, 60, 80, 100]
        print(self.score)

    # 은닉된 함수를 불러내는 함수 (게터)
    def get_score_by_teacher(self):
        self.__get_test_score()

    def get_lecturer_words(self):
        return "Did you prepare well for the lecture?"

    def get_study_site(self, site):
        if site == "classroom":
            return "Off-line"
        if site == "home":
            return "On-line"
```

## 인스턴스 만들기

```python
nlp = GjaiClass("박진수",student_num = 34, class_subject = "NLP")
```
