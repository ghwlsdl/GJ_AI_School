### 파이썬 코드 작성

````python
class Seven_NLP():
  
  Number_of_people = 0
  lecturer = "박진수"
  teacher = "최영화"
  Essential_Equipment_for_Deep_Learning = "Laptop_with_high-end_GPU"
  goal = "Let's develop AI development capabilities and get a job at the company you want!"

  def Main_Teacher(self):
      return "Did you prepare well for the lecture?"

  def Student(self):
      return "To study AI at.광주인공지능사관학교"

  def Study_Site(self, site):
    if site == "classroom":
      return "Off-line"
    if site == "home":
      return "On-line"

  def Attention_Degree(self, level):
    self.level = level
    return self.level
  
  def Attention_Change(self):
    if self.level >= 50:
      self.level -= 10
      return "active Coding"      
    else:
      self.level +=10
      return "need rest"


  def Laptop_Gpu(self, rtx_number):
    
    self.rtx_number = rtx_number

    if  rtx_number > 3060:
      return "Good"
    else:
      return "Not Enough"
````
