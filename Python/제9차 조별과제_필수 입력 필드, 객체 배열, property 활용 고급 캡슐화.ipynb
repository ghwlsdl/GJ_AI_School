### 필드 구분하기

1. 필수필드 : 학생 이름, 강의장소

2. 비필수 필드 : 성별

### 1. class 생성

```python
class NLP_7():
    def __init__(self,
                 name,
                 class_site,
                 sex = ""
                ):
      
        self.name = name
        self.sex = sex
        self.class_site = class_site

    def get_study_site(self, site):
        if site == "classroom":
            return "Off-line"
        if site == "home":
            return "On-line"
```

```python
Hojin = NLP_7(name = "호진", sex = "남", class_site = "home")
```

```python
Seoyeon = NLP_7("서연", "home")
```

```python
Juyoung = NLP_7(name = "주영", class_site = "classroom")
```

### 2. 객체 리스트 생성

```python
NLP_7_list = list()
```

### 3. 객체 리스트에 새로운 객체를 추가

```python
NLP_7_list.append(Hojin)
```

```python
NLP_7_list.extend([Seoyeon,Juyoung])
```

```python
NLP_7_list
```

### 4. 객체 리스트 for 문을 사용해서 출력

```python
for student_name in NLP_7_list:
    print(student_name.get_study_site(student_name.class_site))
```

```python
for instance in NLP_7_list:
  if instance.sex == "남":
    print(f"{instance.name}은(는) 남성 입니다.")
  else:
    print(f"{instance.name}은(는) 여성 입니다.")
```
