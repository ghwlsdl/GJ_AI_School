# **제4차 조별 과제**
---
* 구글 공학계산기의 오른쪽 사칙연산 버튼, sin, cos, tan, log 버튼에 제공하는 기능을 그대로 구현하는 모듈을 만드시오.
* 모듈 이름은 '공학계산기.ipynb'여야 합니다.
* 설계 시에는 다음 요건을 지켜야 합니다.
  * 의사코드를 사용해 모듈과 호출 프로그램을 설계하시오. 
  * 의사코드는 3단계로 나눠 작성해야 합니다. 
  * 1단계는 자연어에 최대한 가깝게, 3단계는 코드에 최대한 가깝게, 2단계는 그 중간 수준으로 작성하시오.
* 모듈 코드를 작성할 때는 다음 요건을 지켜야 합니다.
  * 모듈에 내장할 각 함수의 이름에는 구애 받지 마시오. 
  * 각 함수의 파라미터 이름도 구애 받지 마시오.
  * 파라미터에는 모두 기본값을 지정해 두어야 합니다. 
* 모듈 호출 프로그램(이름은 '호출.ipynb')의 입출력과 관련해서는 다음 요건을 지켜야 합니다. 
  * 입력은 실수 형식으로만 받고, 출력도 실수 형식으로 하시오.
  * 출력 방법에는 제한이 없습니다.
  * 키보드에서 두 가지 숫자를 모두 입력 받아야 합니다. 
  * 필요한 경우에는 두 숫자 중에 더 큰 수를 사용해야 합니다. 
    * 예를 들어 sin(x)의 경우에 x값으로 입력 받은 값 중에 더 큰 값을 사용하는 꼴입니다. 
* 모듈 호출 프로그램의 전반적인 흐름은 다음과 같아야 합니다.
  * 숫자 두 개를 키보드로 입력 받는다.
  * 모듈이 제공하는 각 함수를 호출해 그 계산 결과를 출력한다.
  * 나머지 흐름은 자유재량에 따라 처리하면 됩니다. 
* 기타 요건
  * 나머지 사항은 모두 여러분의 재량껏 하시면 됩니다.
    * 예를 들어, 로그의 밑수로 무엇을 쓸지도 여러분이 알아서 정하세요.

> 작성할 문서(코드) 목록: 반드시 다음 순서대로 문서(노트북 파일 포함)를 작성하셔야 합니다. 

* 의사코드1.ipynb (자연어에 가까운 의사코드)
* 의사코드2.ipynb (자연어와 파이썬 코드 중간 수준 의사코드)
* 의사코드3.ipynb (파이썬 코드에 가까운 의사코드)
* 공학계산기.ipynb (모듈 프로그램)
* 호출.ipynb (호출 프로그램)
