```python
# 라이브러리 math를 import한다.

import math as m


# error = error 변수를 만들고 초기값을 지정한다.

error = "error"


# 사칙연산

def add(x, y):
  return x + y

def sub(x, y):
  return x - y

def mul(x, y):
  return x * y

def div(x, y):
  if y == 0:
    return error
  else:
    return x / y


# 삼각함수

def sin(x, y):
  if x <= y:
    return m.sin(y)
  elif x > y:
    return m.sin(x) 
  else:
    return error

def cos(x, y):
  if x <= y:
    return m.cos(y)
  elif x > y:
    return m.cos(x) 
  else:
    return error

def tan(x, y):
  if x <= y:
    return m.tan(y)
  elif x > y:
    return m.tan(x)
  else:
    return error


# 로그함수
# 로그함수 조건 : 밑은 0보다 크고, 1이면 안 되며, 진수는 0보다 커야 한다.

def log(x, y): # x = 밑, y = 진수
  if x <= 0:
    return error
  elif x == 1:
    return error
  elif y <= 0:
    return error
  else:
    return "앞에 있는 숫자가 밑입니다: ", m.log(y, x)
```
