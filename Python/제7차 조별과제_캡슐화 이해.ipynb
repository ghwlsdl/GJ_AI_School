## **리팩토링 및 변수 은닉**
- 관점 맞추기
- 이름들 적절하게 수정하기

```python
class Seven_NLP():

    def __init__(self):
        self.__lecturer = "박진수"
        self.__lecturer_num = "010-1111-1111"
        self._teacher = "최영화"


    # 은닉된 변수의 값을 불러내는 함수 (게터)
    def get_hided_information(self, title):
        if title == "name":
            print(self.__lecturer)
        elif title == "number":
            print(self.__lecturer_num)
        else:
            print("error")

    # 은닉된 함수
    def __get_test_score(self):
        self.score = [50, 60, 80, 100]
        print(self.score)

    # 은닉된 함수를 불러내는 함수 (게터)
    def get_score_by_teacher(self):
        self.__get_test_score()


    def get_lecturer_words(self):
        return "Did you prepare well for the lecture?"

    def get_study_site(self, site):
        if site == "classroom":
            return "Off-line"
        if site == "home":
            return "On-line"
```

## **캡슐화 테스트하기**

```python
sn = Seven_NLP()
```

> __로 은닉된 변수는 그냥 값을 얻으려고 하면 오류가 발생한다.

```python
sn.__lecturer
```

> 함수를 써야 숨겨진 정보를 얻을 수 있다.


```python
sn.get_hided_information("name")
sn.get_hided_information("number")
```
