> **[과제 내용]**
* 예시 코드를 참조해서 다음과 같이 합니다. 
  * 여러분이 9차 조별 과제에서 만든 클래스를 수정해 생성자에 **가변 위치 인자**(\*args)와 **가변 키워드 인자**(\*kwargs)가 있게 하시오.
  * 클래스를 사용해 인스턴스를 20개 이상 만들어 봅니다.
    * 즉, **[1] 정규 위치 인자**, **[2] 가변 위치 인자**, **[3] 정규 키워드 인자**, **[4] 가변 키워드 인자**들을 넣기도 하고 빼기도 하는 식으로 다양하게 인스턴스를 생성해 봅시다.
    * 에러가 출력되면 출력되는 대로 두고, 왜 에러가 나왔는지를 분석해 보고 그 결과를 새로운 텍스트 셀을 만들어 기록해 둡시다.



```python
class Weather():
    def __init__(self,
                 month,
                 date,
                 place,
                 
                 *weather_detail,

                 condition="",

                 **sun

                ):
        
        self.month = month
        self.date = date
        self.place = place
        self.condition = condition

        self.weather_detail = weather_detail

        self.sunrise_hour = sun["sunrise_h"]
        self.sunrise_min = sun["sunrise_m"]
        self.sunset_hour = sun['sunset_h']
        self.sunset_min = sun['sunset_m']


    def basic_weather_information(self):
        if len(self.condition) != 0:
            print(f"{self.month : >2d}월 {self.date : >2d}일의 \033[1m{self.place}\033[0m의 날씨는 '{self.condition}'입니다.")
        else:
            print(f"{self.month : >2d}월 {self.date : >2d}일의 \033[1m{self.place}\033[0m의 날씨입니다.")
    
    def weather_information_detail(self):
        print(f"오늘의 온도는 {self.weather_detail[0]}도 이고, 습도는 {self.weather_detail[1]}% 입니다.")

    def sun_movement(self):
        print(f"\n오늘의 일출 시간 : {self.sunrise_hour}시 {self.sunrise_min}분, \n오늘의 일몰 시간 : {self.sunset_hour}시 {self.sunset_min}분\n\n")
```

```python
weather_detail = (31.9, 58)
gj = Weather(7,1,"광주", 
             *weather_detail, 
             **{"sunrise_h":5, "sunrise_m":21, "sunset_h":19, "sunset_m":51})
```

```python
gj.basic_weather_information()
gj.weather_information_detail()
gj.sun_movement()
```

```python
seoul.basic_weather_information()
seoul.weather_information_detail()
seoul.sun_movement()
```
