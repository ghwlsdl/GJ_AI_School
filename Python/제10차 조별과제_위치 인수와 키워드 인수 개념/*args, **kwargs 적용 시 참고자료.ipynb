# *args 란?

*args는 *arguments의 줄임말입니다.

하지만, 꼭 저 단어를 쓸 필요는 없습니다.

*a 라고 써도 되고, *kdfsjladshjl 라고 적어도 되고, *myNames라고 적어도 됩니다.

이 지시어는 여러 개(복수 개의)의 인자를 함수로 받고자 할 때 쓰입니다.

예를 들어, 사람의 이름을 받아서 성과 이름을 분리한 후 출력을 하고 싶습니다.

그런데 문제가 생겼습니다.

사용자가 '몇 개의' 이름을 적어 넣을지 알 수가 없다는 것이죠.

이럴 때 *args를 인자로 받아서 사용합니다.

```python
def lastName_and_FirstName(*Names):
    for name in Names:
        print("%s %s" % (name[0], name[1:3]), end=' ')
    print("\n")

lastName_and_FirstName('임서연')
lastName_and_FirstName('임서연', '김재은')
lastName_and_FirstName('임서연', '김재은', '하지성')
lastName_and_FirstName('임서연', '김재은', '하지성', '최아름')
```

# **kwargs 란?

kwargs 는 keyword argument 의 줄임말로 키워드를 제공합니다.

```python
def introduceEnglishName(**kwargs):
    for key, value in kwargs.items():
        print("{0} is {1}".format(key, value))

introduceEnglishName(MyName = 'Hojin')
```

**kwargs는 (키워드 = 특정 값) 형태로 함수를 호출할 수 있습니다.

그것은 그대로 딕셔너리 형태로 {'키워드' : '특정 값'} 요렇게 함수 내부로 전달됩니다.

그렇게 전달받은 딕셔너리를 마음대로 사용하면 되겠죠?

이것을 또 다르게 사용하는 방법으로는

특정 키워드에 반응하여 함수를 작성하는 방법이 있습니다.

예를 들어 다른 이름이 입력 되었을 때 다르게 반응하고 싶게 만들어 봅시다.

```python
def introduceEnglishName(**kwargs):
    for key, value in kwargs.items():
        if 'answer' in kwargs.keys():
            print("Nice to meet you")
        else: print("{0} is {1}".format(key, value))

introduceEnglishName(MyName = 'Hojin')
introduceEnglishName(answer = 'Seoyeon')
```

```python
def blog_printer(*name, blogs):
    print(name)
    for post in blogs:
        print(post)

host = "내 이름은 블로그 주인"
블로그1 = '첫번째 블로그'
블로그2 = '두번째 블로그'
블로그3 = '세번째 블로그'

blog_printer(host, 블로그1, 블로그2, 블로그3)
```

오류가 납니다!

왜냐하면, *arg는 일반 변수보다는 뒤에 위치해있어야 합니다.

파이썬은 어디서부터 어디까지를 *변수에 담을지 알 수 없습니다.

그래서 맨 앞에 특정 변수를 명시해두고, 그 뒤에는 *args로 arguments를 넣어줘야 합니다.

# *arg, **kwargs를 둘 다 써보자.

```python
def blog_printer(name, *blogs, **blog_benefits):
    '''
    name : 블로그 주인장의 이름
    *blogs : 블로그를 만들 때 설명
    **blog_benefits : 해당 블로그의 수익
    '''
    print(name)
    for post in blogs:
        print(post)
    for blog, benefits in blog_benefits.items():
        print(blog, '수익은 >>', benefits)

host = "내 이름은 블로그 주인"
블로그1 = '첫번째 블로그'
블로그2 = '두번째 블로그'
블로그3 = '세번째 블로그'

help(blog_printer)
blog_printer(host, 블로그1, 블로그2, 블로그3, 블로그수익1 = 30, 블로그수익2 = 40, 블로그수익3 = 100)
```
