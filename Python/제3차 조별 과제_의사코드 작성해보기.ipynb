# [제3차 조별 과제]

> 제2차 조별 과제 내용은 다음과 같았습니다.

```
* 우리반 학우들 중에 온라인 반 학우 세 명, 오프라인 반 학우 세명의 이름을 서로 물어보아 알아내서, 
* 그 데이터를 바탕으로 리스트, 튜플, 셋, 딕셔너리 형태로 표현하고, 그 결과를 출력하는 프로그램을 작성하시오. * (단, 출력 모양에는 구애 받지 않아도 됩니다.)
```

* 이제 여러분은 이 문제를 풀면서 작성한 코드를 코드 형식이 아닌 의사코드 형식으로 옮겨 보세요. 즉, 파이썬 형식이 드러나지 않게 최대한 우리말에 가깝게 작성해야 합니다. 

* 두 사람이 한 조가 되어서 이 의사코드를 어떻게 하면 간결하면서도 프로그램의 처리 절차를 잘 표현할 수 있게 의사코드를 작성할 수 있는지  협의해서 작성하시오. 

* 이런 식으로 의사코드를 세 개 이상 작성해서 공유 폴더 안 '제3차 조별과제' 폴더 안에 여러분의 조원 이름이 모두 담긴 폴더를 만들고 그 안에 의사코드 파일을 넣어 두시오.

* 의사코드는 반드시 코랩 노트북의 텍스트셀 안에 작성해야 합니다.

* 각 의사코드 파일 안에는 코드 셀을 먼저 두어 여러분이 작성했던 코드를 복사해서 거기에 두고, 그 다음 자리에 텍스트 셀을 만들어 텍스트 셀 안에 의사코드를 세 개 작성해 두면 됩니다. 


----------


# [원본 코드]

````python
from time import time, sleep

start = time()

online_student_list = ['김재은', '하지성', '최아름']
online_student_tuple = ('김재은', '하지성', '최아름')
online_student_dic = {'김재은' : '온라인' , '하지성' : '온라인', '최아름': '온라인'}

offline_student_list = ['주성민', '박지호', '이찬영']
offline_student_tuple = ('주성민', '박지호', '이찬영')
offline_student_dic = {'주성민' : '오프라인', '박지호' : '오프라인', '이찬영' : '오프라인'}

onsl = online_student_list
onst = online_student_tuple
onsd = online_student_dic

offsl = offline_student_list
offst = offline_student_tuple
offsd = offline_student_dic

print(onsl, onst, onsd, offsl, offst, offsd, sep='\n')

end = time()

print('time elapsed:', end - start)
````
