### loc, iloc 인덱서
- loc : 실제 인덱스 이름 or 컬럼명을 사용해서 가져 올 떄 사용
- iloc : 인덱스 번호를 가져 올 때 사용

```python
df2.loc[:, '2010':'2005']
```

```python
# 부산~인천, 2020~2010 슬라이싱
df2.loc['부산':'인천', '2020':'2010'] 
```

```python
# 부산~인천, 2020~2010 슬라이싱 - Researcher Choi code
df2.iloc[1:3 , 0:2] 
```

```python
# 특정연도에만 존재하고 있는 경찰청 삭제하기 (drop)
data2021 = data2021.drop("광주지방경찰청")
```

```python
data2021
```

```python
# 2. 연도별 데이터 총합 구하기 ('총합'이라는 새로운 column 추가해서 데이터 대입)
data2019['총합'] = data2019.sum(axis = 1)
data2020['총합'] = data2020.sum(axis = 1)
data2021['총합'] = data2021.sum(axis = 1)
```

```python
data2019
```

```python
data2020
```

```python
data2021
```

```python
# 3. 구분이 '발생건수' 인 값들만 가져오기 ('구분' == '발생건수') <- 불리안 활용
temp2019 = data2019[data2019['구분'] == '발생건수']
temp2020 = data2020[data2020['구분'] == '발생건수']
temp2021 = data2021[data2021['구분'] == '발생건수']
```

```python
temp2019
```

```python
temp2020
```

```python
temp2021
```

```python
#총합을 기준으로 데이터 잘라오기
s2019 = temp2019.loc[:, '총합']
s2020 = temp2020.loc[:, '총합']
s2021 = temp2021.loc[:, '총합']
```

```python
s2019
```

```python
s2020
```

```python
s2021
```

```python
s2019.name = "2019 총계"
s2020.name = "2020 총계"
s2021.name = "2021 총계"
```

```python
s2019
```

```python
s2020
```

```python
s2021
```

```python
#증감률 계산
# (금년 - 작년) / 작년 * 100

plus19_20 = (s2020 - s2019) / s2019 * 100
plus19_20.name = "2019-2020 증감률"
plus20_21 = (s2021 - s2020) / s2020 * 100
plus20_21.name = "2020-2021 증감률"
```

```python
print(plus19_20)
print(plus20_21)
```

```python
# 데이터프레임으로 병합하기 (concat)
pd.concat([s2019, plus19_20, s2020, plus20_21, s2021], axis = 1)
```
