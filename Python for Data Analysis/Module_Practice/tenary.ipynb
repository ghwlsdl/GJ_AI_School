> **[과제 내용]**
  * 학습용 교재인 '파이썬 라이브러리를 활용한 데이터 분석'의 2장에 나온 코드를 참고하여 다음과 같은 모듈을 따로 만들고, 이를 불러오거나(import) 매직 명령어로 실행해(%run) 그 실행 결과를 출력하는 프로그램을 작성합시다. 즉, 모듈 한 개 + 실행 프로그램 한 개 꼴이어야 합니다.
  * 모듈은 1개 함수로 되어 있어야 합니다.
    * 모듈의 함수는 다음과 같은 네 개 값을 받아야 합니다.
      * [필수 필드 1] 고정 길이 위치 인자: 반 번호, 담임 이름, 부담임 이름
      * [필수 필드 2] 가변 길이 위치 인자(args): 학생 이름들
      * [선택 필드 1] 고정 길이 키워드 인자: 정원, 출석학생수, 결석학생수, 지각학생수
      * [선택 필드 2] 가변 길이 키워드 인자: 학생별 형성평가성적을 학생 이름으로 키워드를 삼습니다. 
    * 모듈의 함수는 다음과 같은 값들을 반환함과 동시에 출력해 주어야 합니다(단, 출력 모양에는 구애 받지 않아도 됩니다).
      * [1] 반 번호, 담임 이름, 부담임 이름
      * [2] 출석률, 결석률, 지각률(단, 지각률은 0.1로 합니다.)
      * [3] 학생들의 평균 파이썬 성적
    * 모듈의 이름은 tenary.ipynb여야 합니다.
  * 실행 프로그램에서는 다음과 같이 해야 합니다.
    * (1) 모듈을 가져와서(import)해서 반환 값들을 활용해 모듈에서 출력하는 방식과 같은 식으로 출력해 봅니다. 
    * (2) 매직 명령어를 사용해 상기 파이썬 노트북을 실행하고 그 결과로 반환 받은 값들을 사용해 내부 관찰(introspection) 방법으로 확인하고 모듈에서 출력하는 방법과 동일하게 출력해 주어야 합니다(단, 출력 모양에는 구애 받지 않아도 됩니다).
  * 이럼으로써 세 개의 출력 결과를 얻을 수 있어야 합니다.
    * (1) 모듈 자체의 출력
    * (2) import한 모듈을 이용한 출력
    * (3) 매직 명령어로 가져온 값을 이용한 출력

2. 실행
  - 구글 드라이브 마운트
  - 필요한 라이브러리 import
  - 매직 명령어로 실행

```python
import random

def class_information_control(self,
          
                              # 필수 필드
                              class_nums, #반 번호
                              lecturer_name, #담임 이름
                              sub_teacher_name, #부담임 이름


                              # 가변 길이 위치 인자 (*args)
                              *students_name, # 시험을 본 학생


                              # 비필수 필드
                              number_of_students = 0, # 정원
                              number_of_tested_students = 0, # 시험을 본 학생 수
                              number_of_absent_students = 0, # 결석한 학생 수
                              number_of_late_students = 0, # 지각한 학생 수


                              # 가변 길이 키워드 인자 (**kwargs)
                              **score_of_students # 시험을 본 학생 성적 
                            ):


                              # 출석한 학생 수, 결석한 학생 수
                              number_of_tested_students = len(students_name)
                              number_of_absent_students = number_of_students - number_of_tested_students


                              # 출석률, 결석률, 지각률
                              rate_of_number_of_tested_students = number_of_tested_students / number_of_students * 100
                              rate_of_absent_students = number_of_absent_students / number_of_students * 100
                              rate_of_late_students = number_of_late_students / number_of_students * 100


                              # 학생들의 평균 성적 구하기
                              score_sum = 0

                              for score in score_of_students.values():
                                  score_sum += score

                              score_average = score_sum / number_of_tested_students # 전체 성적 / 시험을 본 학생 수

                              # (반 번호, 담임 이름, 부담임 이름) 을(를) 출력
                              print(f'''
                              반 번호 : {class_nums}
                              강사 : {lecturer_name}
                              부담임 : {sub_teacher_name}
                              ''')


                              # (시험을 본 학생, 출석한 학생 수, 결석한 학생 수, 지각한 학생 수) 을(를) 출력
                              print(f'''
                              시험을 본 학생 : {students_name}
                              출석한 학생 수  : {number_of_tested_students}
                              결석한 학생 수 : {number_of_absent_students}
                              지각한 학생 수 : {number_of_late_students}
                              ''')
                              

                              # (출석률, 결석률, 지각률) 을(를) 출력한다.
                              print(f'''
                              출석률 : {rate_of_number_of_tested_students:.2f} %
                              결석률 : {rate_of_absent_students:.2f} %
                              지각률 : {rate_of_late_students:.2f} %
                              ''')


                              # 평균 파이썬 성적을 출력한다.
                              print(f'''
                              학생들 평균 성적 : {score_average} 점
                              ''')
```
